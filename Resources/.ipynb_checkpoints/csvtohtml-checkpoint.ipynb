```python
#import dependencies
import pandas as pd
import os
```

```python
#load in csv file and have pandas read it
file=os.path.join('cities.csv')
city=pd.read_csv(file, header=0, index_col=False)
```

```python
#create df 
data_df=pd.DataFrame(city)
data_df
```

```python
#add City ID to index column
data_df.columns.name='City ID'
data_df
```

```python
data_df.to_html('cities.html')
```
